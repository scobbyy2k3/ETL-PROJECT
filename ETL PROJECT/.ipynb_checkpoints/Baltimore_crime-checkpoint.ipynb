```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import csv
from scipy import stats
```

```python
path = "911_calls_for_service.csv"
baltimore__911_df = pd.read_csv(path, low_memory=True, nrows=5000 )
baltimore__911_df.head()
```

```python
baltimore__911_df = baltimore__911_df.drop(columns=['Unnamed: 0'])
baltimore__911_df.head()
```

```python
# type = baltimore__911_df["district"].value_counts()
# type
```

```python
baltimore__911_df.rename(columns={'district':'District'}, inplace=True)
baltimore__911_df.head()
```

```python
path = "BPD_Part_1_Victim_Based_Crime_Data.csv"
baltimore__crime_df = pd.read_csv(path, low_memory=True, nrows=5000  )
baltimore__crime_df.head()
```

```python
#extract year
baltimore__crime_df['year'] = pd.DatetimeIndex(baltimore__crime_df['CrimeDate']).year
baltimore__crime_df
```

```python
# types = baltimore__crime_df["year"].value_counts()
# types
```

```python
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("NORTHEASTERN","NE")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("SOUTHEASTERN","SE")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("SOUTHERN","S")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("NORTHERN","N")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("CENTRAL","C")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("NORTHWESTERN","NW")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("SOUTHWESTERN","SW")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("EASTERN","E")
baltimore__crime_df["District"]= baltimore__crime_df["District"].replace("WESTERN","W")
```

```python
baltimore__crime_df
```

```python
merge_table = pd.merge(baltimore__911_df, baltimore__crime_df, on="District")
merge_table.head()
```

## merge_table.columns

```python
merge_table = pd.DataFrame({
    "CrimeDate":merge_table["CrimeDate"],
    "CrimeTime":merge_table["CrimeTime"],
    "District":merge_table["District"],
    "Description":merge_table["Description"],
    "Weapon":merge_table["Weapon"],
    "Total Incidents":merge_table["Total Incidents"]
})
merge_table.head()
```

```python
merge_table = merge_table.dropna()
merge_table.head()
```
